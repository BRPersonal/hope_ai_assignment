```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import pandas as pd
```

```python
# 1. Define our classes
classes = ['Cat', 'Dog', 'Bird']

# 2. Create "Actual" labels (the ground truth)
# Let's assume 30 Cats, 40 Dogs, and 30 Birds
y_true = (['Cat'] * 30) + (['Dog'] * 40) + (['Bird'] * 30)

# 3. Create "Predicted" labels (simulating model errors)
# We'll intentionally swap some to see the False Positives/Negatives
y_pred = (['Cat'] * 25 + ['Dog'] * 3 + ['Bird'] * 2) + \
         (['Dog'] * 35 + ['Cat'] * 3 + ['Bird'] * 2) + \
         (['Bird'] * 28 + ['Cat'] * 1 + ['Dog'] * 1)

# 4. Generate the Matrix
cm = confusion_matrix(y_true, y_pred, labels=classes)
```

```python
# 5. Plotting using Seaborn for better readability
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=classes, yticklabels=classes, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Multi-Class Confusion Matrix')
plt.show()
```
