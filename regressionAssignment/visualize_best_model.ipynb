```python
#CELL-NO: 1
import pandas as pd
import pickle
import warnings
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
#suppress warnings
warnings.filterwarnings('ignore')
print("imports successful")
```

```python
#CELL-NO: 2
filename = "insurance_model.sav"
loaded_model = pickle.load(open(filename,"rb"))
print("model loaded")
```

```python
#CELL-NO: 3
full_data = pd.read_csv("insurance_pre.csv")
data = full_data.head(20) #let us visualize first twenty rows alone
#show top 5 rows
data.head()
```

```python
#CELL-NO: 4
data.shape
```

```python
#CELL-NO: 5
data = pd.get_dummies(data,drop_first=True)
data.head()
```

```python
#CELL-NO: 6
output_col = "charges"
x = data.iloc[:,data.columns != output_col]
y = data.iloc[:,data.columns == output_col]
```

```python
#CELL-NO: 7
x.head()
```

```python
#CELL-NO: 8
y.head()
```

```python
#CELL-NO: 9
_,x_test,_,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
y_pred = loaded_model.predict(x_test)
```

```python
#CELL-NO: 10
#plotting
plt.plot(x_test,y_test,"*",color="green")
plt.plot(x_test,y_pred,"+",color="red")
plt.title("Actual vs Predicted")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```
